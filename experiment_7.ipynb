```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df = pd.read_csv("data4.csv")
print("Original Dataset:")
print(df.head())

num_cols = df.select_dtypes(include='number').columns

df[num_cols].hist()
plt.suptitle("Before Scaling")
plt.show()

cat_cols = df.select_dtypes(include='object').columns

if len(cat_cols) > 0:
    df_encoded = pd.get_dummies(df, columns=cat_cols)
else:
    df_encoded = df.copy()

minmax = MinMaxScaler()
df_minmax = df_encoded.copy()
df_minmax[num_cols] = minmax.fit_transform(df_minmax[num_cols])

print("\nAfter Min-Max Scaling:")
print(df_minmax.head())

df_minmax[num_cols].hist()
plt.suptitle("After Min-Max Scaling")
plt.show()

zscore = StandardScaler()
df_zscore = df_encoded.copy()
df_zscore[num_cols] = zscore.fit_transform(df_zscore[num_cols])

print("\nAfter Z-Score Normalization:")
print(df_zscore.head())

df_zscore[num_cols].hist()
plt.suptitle("After Z-Score Normalization")
plt.show()
```
