```python
import pandas as pd

df = pd.read_csv('data4.csv')

print("Shape (Rows, Columns):", df.shape)

print("\nFirst 5 Rows:")
print(df.head())

print("\nSize of dataset:", df.size)

print("\nMissing Values:")
print(df.isnull().sum())

num_df = df.select_dtypes(include='number')

print("\nSum:")
print(num_df.sum())

print("\nMean:")
print(num_df.mean())

print("\nMinimum values:")
print(num_df.min())

print("\nMaximum values:")
print(num_df.max())
```
