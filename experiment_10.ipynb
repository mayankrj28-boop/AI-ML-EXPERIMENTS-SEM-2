```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("data10.csv")

print(df.head())

# MATPLOTLIB GRAPHS

# Line Plot
plt.plot(df["Age"], df["Daily_Usage_Hours"])
plt.title("Age vs Daily Usage")
plt.xlabel("Age")
plt.ylabel("Usage Hours")
plt.show()

# Bar Chart
plt.bar(df["Age"], df["App_Downloads"])
plt.title("Age vs App Downloads")
plt.xlabel("Age")
plt.ylabel("Downloads")
plt.show()

# Histogram
plt.hist(df["Screen_Time"])
plt.title("Screen Time Distribution")
plt.show()

# Scatter Plot
plt.scatter(df["Daily_Usage_Hours"], df["Battery_Drain"])
plt.xlabel("Usage Hours")
plt.ylabel("Battery Drain")
plt.title("Usage vs Battery")
plt.show()

# SEABORN GRAPHS

# Boxplot
sns.boxplot(x=df["Screen_Time"])
plt.title("Boxplot of Screen Time")
plt.show()

# Heatmap (Correlation)
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Pairplot (VERY IMPRESSIVE)
sns.pairplot(df)
plt.show()
```
